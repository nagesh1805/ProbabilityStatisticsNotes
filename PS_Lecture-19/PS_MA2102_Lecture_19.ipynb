**Theorem:** Let $X$ is a random variable with $CDF$ $F_X(x)$, then $Y=F_X(X)\sim U[0,1]$

*proof:* $CDF$ of $Y$, $S_Y=[0,1]$ &nbsp; &nbsp;($\because$ $0\le F_X(x)\le 1$)

for $0\le y\le1$, $F_Y(y)=P(Y\le y)$

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; $=P(F_X(X)\le y)$

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; $=P(X\le F_X^{-1}(y))$&nbsp; &nbsp; ($\because$ $F_X$ is $\nearrow$ $\Rightarrow$ $F_X^{-1}$$\nearrow$)

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; $=F_X(F_X^{-1}(y))=y$

for $y<0$, $F_Y(y)=P(Y\le y)=P(F_X(X)\le y)=0$ &nbsp; &nbsp;($\because$ $F_X(x)\ge 0$)

for $y> 1$, $F_Y(y)=P(Y\le y)=P(F_X(X)\le y)=1$ &nbsp; &nbsp;($\because$ $F_X(x)\le 1$)

$\therefore$ $F_Y(y)=\left\{\begin{matrix}
 0&y<0 \\ 
 y&0\le y\le1 \\
 1&y>1
\end{matrix}\right.$

$\therefore$ $Y\sim U[0,1]$

**Theorem:** Let $X$ is a continuous random variable with $CDF$ $F_X(x)$ & let $U$ be the uniform random variable over $[0,1]$ then $Y=F_X^{-1}(U)$ is a random variable with $CDF$ $F_X$

*proof:* Here $U\sim U[0,1]$, $F_U(u)=\left\{\begin{matrix}
 0&u<0 \\
 u&0\le u\le 1 \\
 1& u>1
\end{matrix}\right.$

$Y=F_X^{-1}(U)$

$CDF$ of $Y$, $F_Y(y)=P(Y\le y)$

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;  &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; $=P(F_X^{-1}(U)\le y)=P(U\le F_X(y))$

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;  &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; $=F_U(F_X(y))$

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;  &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; $=F_X(y)$ &nbsp; &nbsp; &nbsp;($\because~0\le F_X(y)\le 1$)

$\therefore$ $F_Y(y)=F_X(y)$ $\forall y\in \mathbb{R}$

Hence $Y$ has the  $CDF$ $F_X$

**Note:** If $u_1,u_2,...,u_n$ are realizations of random variable $U\sim U[0,1]$, then $F_X^{-1}(u_1),F_X^{-1}(u_2),...F_X^{-1}(u_n)$ will be realizations of random variable $X$ whose $CDF$ is $F_X$. So this is how one can generate data related to any distribution once we have access to uniform data in $[0,1]$

**Example 1:** $X\sim Exp(\lambda)$ $\Rightarrow$ $F_X(x)=1-e^{-\lambda x}$, $x>0$

Let $U\sim U[0,1]$ then $Y=F_X^{-1}(U)=-\frac{1}{\lambda}\log(1-U)\sim Exp(\lambda)$

**Example 2:** Let $X\sim Rayleigh(b)$ where $b>0$ is parameter of distribution

$CDF$ of $X$, $F_X(x)=1-e^{-\frac{b}{2}x^2}$, $x>0$, and let $U\sim U[0,1]$

$Y=F_X^{-1}(U)=\sqrt{-\frac{2}{b}\log(1-U)}\sim Rayleigh(b)$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Infty5 =10

fig, ax = plt.subplots(2,2,figsize=(12, 10))
n=100000
u=np.random.rand(n) # Generating random sample of size n from uniform distribution over [0,1]
def foo1(lmda):
    y1=(-1/lmda)*np.log(1-u) # Generating random sample of size n from Exponetial distribution  with paramter lmda
    ax[0][0].hist(y1,bins=100,density=True)
    ax[0][0].set_title("Histogram of Exponetial Data")
    ax[0][0].set_xlim(0,5)

def foo2(lmda):
    x=np.linspace(0,Infty5,1000)
    ax[1][0].plot(x,lmda*np.exp(-lmda*x))
    ax[1][0].set_title("Graph of Exponential PDF")
    ax[1][0].set_xlim(0,5)
    
def foo3(b):
    y2=np.sqrt((-2/b)*np.log(1-u)) # Generating random sample of size n from Rayleigh distribution  with paramter b
    ax[0][1].hist(y2,bins=100,density=True)
    ax[0][1].set_title("Histogram of Rayleigh Data")
    ax[0][1].set_xlim(0,5)
def foo4(b):
    x=np.linspace(0,Infty5,1000)
    ax[1][1].plot(x,b*x*np.exp(-(b/2)*x**2))
    ax[1][1].set_title("Graph of Rayleigh PDF")
    ax[1][1].set_xlim(0,5)

lmda, b = 4, 3    
foo1(lmda)  # lmda=4
foo2(lmda)  # lmda=4
foo3(b)  #b=3
foo4(b) # b=3
plt.show()
```

# Multivariate Distributions

Thus far, we have considered only probability distributions of single(univariate) random variables, but we are often interested in probability statements concerning two(Bivariate) or more(Multivariate) random variables

Consider the random experiment of rolling two distinct dice. Here the sample space $\Omega=\{(i,j):i,j\in \{1,2,3,4,5,6\}\}$

Let $X_1:\Omega\to\mathbb{R}$, such that $X_1(i,j)=i+j$

$X_2:\Omega\to\mathbb{R}$, such that $X_2(i,j)=ij$

It may be our interest to study numerical characteristics $X_1$, $X_2$ simultaneously. This amounts to study of two variable function $\underline{X}\equiv(X_1,X_2):\Omega\to\mathbb{R}^2$

where $\underline{X}(\omega)=(X_1(\omega),X_2(\omega))$

Next consider the situation where the doctor examining a random patient at Government hospital. To study a patient health a doctor may need to study certain numerical characteristics of patient like $X_1:$ height, $X_2:$ weight, $X_3:$ temperature, and $X_4:$ blood pressure, simultaneously.

This amounts to study of four variable function $\underline{X}\equiv(X_1,X_2,X_3,X_4):\Omega\to\mathbb{R}^4$

**Definition:** For given probability space $(\Omega,\mathscr{F},P)$, a function $\underline{X}\equiv(X_1,X_2,...,X_k):\Omega\to\mathbb{R}^k$ is called $k-$dimensional random vector if $X^{-1}(B)\in\mathscr{F}$ $\forall B\in \mathscr{B}_k$

where $\mathscr{B}_k$ is Borel $\sigma-$Field in $\mathbb{R}^k$ which is generated by $\mathscr{I}_k=\{ (a_1,b_1]\times(a_2,b_2]\times...\times(a_k,b_k]\subset\mathbb{R}^k:a_i,b_i\in\mathbb{R},a_i<b_i \text{ for } i=1,2,...k\}$

and $\mathscr{B}_k$ is also generated by $\mathscr{J}_k=\{ (-\infty,a_1]\times(-\infty,a_2]\times...\times(-\infty,a_k]\subset\mathbb{R}^k:a_i\in\mathbb{R} \text{ for } i=1,2,...k\}$

A 1-dimensional random vector will be referred as a <b>univariate random variable</b>

and 2-dimensional random vector will be referred as <b>bivariate random variable</b>

From now on we restrict attention to two-dimensional random vector, and the discussion for $k-$dimensional ($k>2$) case is similar.

**Note:** Basic events (which generates whole Borel $\sigma-$Field) in $\mathscr{B}_2$ are rectangular regions in plane (bounded or unbounded)

### Joint Probability distribution of bivariate random variable $(X,Y)$:

A function $P^{XY}:\mathscr{B}_2\to\mathbb{R}$, defined as

$P^{XY}(B)=P((X,Y)\in B)=P\{\omega\in\Omega:(X(\omega),Y(\omega))\in B\}$

One can easily check that $P^{XY}$ function satisfies three axioms of probability function. Here $P^{XY}$ is called the joint probability distribution of $(X,Y)$

We don't have much tools to study $P^{XY}$ as domain of $P^{XY}$ is Borel $\sigma-$ Field(set of sets). So we are going to study another two variable function of real variables called <b>Joint cumulative distribution function</b> 

### Joint Cumulative distribution function (JCDF):

Let $(X,Y)$ be bivariate random variable, and a function $F_{X,Y}:\mathbb{R}^2\to\mathbb{R}$ defined by

$F_{X,Y}(x,y)=P^{XY}((-\infty,x]\times(-\infty,y])=P(X\le x, Y\le y) ~~\forall (x,y)\in \mathbb{R}^2$ is called <b>Joint cumulative distribution function ($JCDF$)</b> of $(X,Y)$

As we can see that $JCDF$ has complete information about probabilities of Borel sets of the form $(-\infty,x]\times(-\infty,y]$ and we know sets of this type actually generate whole Borel $\sigma-$ Field $\mathscr{B}_2$ in $\mathbb{R}^2$. Therefore Joint cumulative distribution function uniquely determine the joint probability distribution.

**Properties of JCDF**:

i) $F_{X,Y}(x,y)$ is non-decreasing w.r.t each coordinate.

i.e $F_{X,Y}(x_1,y)\le F_{X,Y}(x_2,y)$ for $x_1<x_2$ ; $F_{X,Y}(x,y_1)\le F_{X,Y}(x,y_2)$ for $y_1<y_2$

ii)  $F_{X,Y}(x,y)$ is right continuous w.r.t each coordinate.

i.e $\lim_{h\to 0+}F_{X,Y}(x+h,y)=F_{X,Y}(x,y)$; $\lim_{k\to 0+}F_{X,Y}(x,y+k)=F_{X,Y}(x,y)$

iii) $\lim_{x\to\infty,y\to\infty}F_{X,Y}(x,y)=F_{X,Y}(\infty,\infty)=1$; $\lim_{x\to -\infty}F_{X,Y}(x,y)=F_{X,Y}(-\infty,y)=0=F_{X,Y}(x,-\infty)=\lim_{y\to -\infty}F_{X,Y}(x,y)$

**Theorem:** A function $F:\mathbb{R}^2\to\mathbb{R}$ is $CDF$ of some bivariate random variable if and only if it satisfies the following conditions.

i) $F$ is non-decreasing and right continuous w.r.t both coordinates.

ii) $F(-\infty,y)=F(x,-\infty)=0$ and $F(\infty,\infty)=1$

The distributions of $X$, and $Y$ can be obtained from joint distribution of $X,Y$

For $B\in \mathscr{B}$, $B\times\mathbb{R}\in\mathscr{B}_2$, and $\mathbb{R}\times B\in \mathscr{B}_2$

$P^X(B)=P(X\in B)=P((X,Y)\in B\times \mathbb{R})=P^{XY}(B\times\mathbb{R})$

$P^Y(B)=P(Y\in B)=P((X,Y)\in \mathbb{R}\times B)=P^{XY}(\mathbb{R}\times B)$

$F_X(x)=P(X\le x)=P(X\le x,Y<\infty)=\lim_{y\to\infty}P(X\le x, Y\le y)=\lim_{y\to\infty}F_{X,Y}(x,y)=F_{X,Y}(x,\infty)$

$F_Y(y)=P(Y\le y)=P(X<\infty,Y\le y)=\lim_{x\to\infty}P(X\le x, Y\le y)=\lim_{x\to\infty}F_{X,Y}(x,y)=F_{X,Y}(\infty,y)$

Here $F_X$ and $F_Y$ are referred to as marginal $CDF$'s of $(X,Y)$, and $P^X$ and $P^Y$ are referred to as marginal probability distributions of $(X,Y)$

**Note**:From given joint probability distribution of $(X,Y)$ we can get its marginal probability distributions , but from given marginal distributions of $(X,Y)$, it is not always possible to get their joint probability distribution.

All joint probability statements about $X$ and $Y$ can (in theory) be answered in terms of their $JCDF$.

**Definition:** A bivariate random variable $(X,Y)$ is said to be of the discrete type if both $X,Y$ are univariate discrete type.

If let $S_X$ be the support of $X$, $S_Y$ be the support of $Y$, then $S_{X,Y}=\{(x,y)\in S_X\times S_Y: P(X=x,Y=y)>0\}$ will be support of $(X,Y)$

**Joint probability mass function (JPMF):**

Let $(X,Y)$ be bivariate discrete random variable with support $S_{X,Y}$, then a function $p_{X,Y}:\mathbb{R}^2\to\mathbb{R}$, defined as <br><br>
$p_{X,Y}(x,y)=\left\{\begin{matrix}
 P(X=x,Y=y)& (x,y)\in S_{X,Y} \\ 
 0 & \text{otherwise} 
\end{matrix}\right.$ <br><br>
is called the Joint probability mass function($JPMF$) of $(X,Y)$

**Properties of JPMF:**

i) $p_{X,Y}(i,j)\ge 0$

ii) $\sum_{(i,j)\in S_{X,Y}}p_{X,Y}(i,j)=1$

- For any $B\in\mathscr{B}_2\subset\mathbb{R}^2$, $P((X,Y)\in B)=\sum_{(x,y)\in B\cap S_{X,Y}}p_{X,Y}(x,y)$

From the $JPMF$ we can obtain marginal $PMF$'s as follows:

 for $i\in S_X$,&nbsp; $p_{X}(i)=P(X=i)=P(X=i,Y\in S_Y)=\sum_{j\in S_Y}P(X=i,Y=j)$

$\therefore~p_X(i)=\sum_{j\in S_Y}p_{X,Y}(i,j)$, for $i\in S_X$

similarly $p_Y(j)=\sum_{i\in S_X}p_{X,Y}(i,j)$ for $j\in S_Y$

Here $p_X$,$p_Y$ are referred to as marginal $PMF$'s of $(X,Y)$

**Note:** In general from marginal $PMF$'s, it is not always possible to obtain their $JPMF$.

**Conditional PMF's:**

 For given $j\in S_Y$ conditional $PMF$ of $X=i$, given $Y=j$ is denoted by $p_{X/Y}(i/j)$ and defined as

$p_{X/Y}(i/j)=P(X=i/Y=j)=\frac{P(X=i,Y=j)}{P(Y=j)}$

$\therefore$ $p_{X/Y}(i/j)=\frac{p_{X,Y}(i,j)}{p_Y(j)}$, for $i\in S_X$

similarly for $i\in S_X$ conditional $PMF$ of $Y=j$ given $X=i$

$p_{Y/X}(j/i)=\frac{p_{X,Y}(i,j)}{p_X(i)}$, for $j\in S_Y$

Here $p_{X/Y}$,$p_{Y/X}$ are called conditional $PMF$'s of $(X,Y)$

**Definition:** A bivariate random variable $(X,Y)$ is said to be of the continuous type if there exists a non-negative function $f_{X,Y}:\mathbb{R}^2\to\mathbb{R}$, such that <br><br>
$F_{X,Y}(x,y)=\int_{-\infty}^{x}\int_{-\infty}^{y}f_{X,Y}(s,t)dtds$ $\forall (x,y)\in \mathbb{R}^2$

Here $f_{X,Y}(x,y)$ is called <b>Joint probability density function (JPDF)</b> of $(X,Y)$

**Properties of JPDF:**

i) $f_{X,Y}(x,y)\ge 0$

ii) $\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}f_{X,Y}(s,t)dtds=1$   &nbsp; &nbsp; ($\because~F_{X,Y}(\infty,\infty)=1$)  (i.e Total volume under the surface of $f_{X,Y}$ is 1)

Support of $(X,Y)$, $S_{X,Y}=\{(x,y)\in \mathbb{R}^2: f_{X,Y}(x,y)>0\}$

- If $f_{X,Y}$ is continuous at $(x,y)$, then $\frac{\partial^2F_{X,Y}(x,y)}{\partial x\partial y}=f_{X,Y}(x,y)$

- For $B\in \mathscr{B}_2\subset\mathbb{R}^2$,&nbsp; $P((X,Y)\in B)=\underset{B}{\int\int}f_{X,Y}(x,y)dxdy$

- We can obtain $PDF$'s of $X,Y$ from $JPDF$ as follows

$F_X(x)=P(X\le x)=P(X\le x,Y<\infty)=\int_{-\infty}^{x}\int_{-\infty}^{\infty}f_{X,Y}(s,t)dtds$

$F_X(x)=\int_{-\infty}^{x}\left(\int_{-\infty}^{\infty}f_{X,Y}(s,t)dt\right)ds$

$f_X(x)=\frac{dF_X(x)}{dx}=\int_{-\infty}^{\infty}f_{X,Y}(x,t)dt$  &nbsp; ($\because \text{First fundemental theorem of Calculus}$)

$\therefore$ $f_X(x)=\int_{-\infty}^{\infty}f_{X,Y}(x,y)dy$

similarly $f_Y(y)=\int_{-\infty}^{\infty}f_{X,Y}(x,y)dx$

Here $f_X$, $f_Y$ are called the Marginal $PDF$'s of $(X,Y)$

## Conditional PDF's:

Conditional $PDF$ of $X$ at $x$ given $Y=y$,  $f_{X/Y}(x/y)=\frac{f_{X,Y}(x,y)}{f_Y(y)}$

Conditional $PDF$ of $Y$ at $y$ given $X=x$,  $f_{Y/X}(y/x)=\frac{f_{X,Y}(x,y)}{f_X(x)}$

**Problem:** You get a stick and break it randomly into three pieces. What is the probability that you can make a triangle using three pieces?

*solution:* Let the length of stick is 1 unit. Let us denote the left end of  stick with 0 and right end with 1

Let $X$, $Y$ denote the location on the stick where we break randomly (from left to right respectively)

then the lengths of three pieces we get are $X$,$Y-X$, and $1-Y$. <br><br>And these pieces make triangle $\iff$ $0<X<\frac{1}{2}$,$0<Y-X<\frac{1}{2}$, and $0<1-Y<\frac{1}{2}$ $\iff$  $0<X<\frac{1}{2}$, and $\frac{1}{2}<Y<X+\frac{1}{2}$<br><br>
($\because$ if one piece length $>\frac{1}{2}$, then the sum of the other two pices will be $<\frac{1}{2}$ as total length is 1 unit)

The $JPDF$ of $f_{X,Y}(x,y)=f_{X}(x)f_{Y/X}(y/x)$

$f_{X,Y}(x,y)=\left\{\begin{matrix}
 \frac{1}{1-0}\times \frac{1}{1-x}=\frac{1}{1-x}& 0<x<y<1 \\ 
 0& \text{otherwise} 
\end{matrix}\right.$                   &nbsp; &nbsp; ($\because$ $X$ is uniformly chosen from 0 to 1 then (after X is chosen) Y is chosen uniformly from X to 1)

$P(\text{ three pieces making a triangle})=P(0<X<\frac{1}{2},\frac{1}{2}<Y<\frac{1}{2}+X)=\underset{\{(x,y):~0<x<\frac{1}{2},~\frac{1}{2}<y<x+\frac{1}{2}\}}{\int\int}f_{X,Y}(x,y)dxdy$

&nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;$=\int_{0}^{\frac{1}{2}}\int_{\frac{1}{2}}^{x+\frac{1}{2}}\frac{1}{1-x}dydx=\int_{0}^{\frac{1}{2}}\frac{1}{1-x}\left(\int_{\frac{1}{2}}^{x+\frac{1}{2}}dy\right)dx=\int_{0}^{\frac{1}{2}}\frac{x}{1-x}dx$

&nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;$=\int_{0}^{\frac{1}{2}}\left(\frac{1}{1-x}-1\right)dx=\left[-\log_e{(1-x)}\right]_0^{\frac{1}{2}}-\left[x\right]_0^{\frac{1}{2}}$

&nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;$=\left[-\log_e{(1-\frac{1}{2})+\log_e{(1-0)}}\right]-[\frac{1}{2}-0]=\log_e{(2)}-1=0.193147$

$\because P(\text{ three pieces making a triangle})=0.193147$

### Simulation of breaking stick into three pieces experiment:

```python
#
# Simulation of breaking stick into three pieces experiment
#

import random

n=1000000
f=0
for _ in range(n):
    x=random.uniform(0,1)
    y=random.uniform(x,1)
    # 0_________X_________Y________1
    # ....x.... ...y-x.... ..1-y...         

    # let three pieces legths are a,b,c then a=x, b=y-x, and c=1-y

    a,b,c = x,y-x,1-y
    if a+b > c and b+c > a and a+c > b :   # if  three pieces form a traingle then f += 1
        f += 1
print("prob = {0}".format(f/n))

```
